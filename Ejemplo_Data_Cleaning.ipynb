```python
import pandas as pd
import numpy as np
```

```python
# Crear un nuevo DataFrame con datos arbitrarios
datos2 =pd.DataFrame(
    {'Columna1':[100,200,300,400],
     'Columna2':[777,123,456,777],
     'Columna3':['Hugo','Paco','Luís','Ana']
     })
datos2
```

```python
# Valores unicos de la columna2
datos2['Columna2'].unique()
```

```python
# Frecuencia para cada valor único de la Columna2
datos2['Columna2'].value_counts()
```

```python
# Crear un nuevo DataFrame con datos arbitrarios
datos3 = pd.DataFrame(
    {'Cédula': ['777', '888', '999', '888'],
     'Salario': [2500, 3400, 1600, 3400],
     'Nombre': ['Hugo', 'Paco', 'Luís', 'Paco']
     })
datos3
```

```python
# Detección de valores duplicados con el método (duplicated)
datos3.duplicated()
```

```python
# Calcular la cantidad de registros duplicados en el DataFrame
datos3.duplicated().sum()
```

```python
# Eliminar los registros o filas duplicadas
datos3.drop_duplicates()
```

```python
# si cargamos el DataFrame el valor duplicado sigue en el conjunto de datos
datos3
```

```python
# Eliminar los registros o filas duplicadas y reemplazar en memoria
datos3.drop_duplicates(inplace=True)
datos3
```

```python
# Crear un nuevamente el DataFrame
datos3 = pd.DataFrame({'Cédula': ['777', '888', '999', '888'], 'Salario': [2500, 3400, 1600,3400], 'Nombre': ['Hugo', 'Paco', 'Luís', 'Paco']})
datos3.drop_duplicates(['Cédula'])
```

### Tratamiento de valores atípicos (outliers)

```python
# Crear un DataFrame con la variable edad
datos4 = pd.DataFrame({'Edad':[151,16,25,34,52,43,70,61,7,79]})
datos4
```

```python
#Reemplazar los valores atípicos con el límite más cercano
datos4.clip(lower=18,upper=80)
```
